```python
import pandas as pd
from functools import reduce

dfs = [eco_active_df,eco_inactive_df,eco_ocup_df,eco_desocup_df]

actividades_final = reduce(lambda left,right:pd.merge(left,right,on=['ENTIDAD','NOM_ENT','MUN','NOM_MUN','AGEB','POBTOT']),dfs)
actividades_final
```
