```python
 # Dependencies
import requests
import json
import pandas as pd

# Google developer API key
from config_maps import gkey
```

```python
#Base de Longuitud y Latitud
latlng_pd = pd.read_csv("coordenadas.csv")
latlng_pd.head()
#latlng_pd['Number of near Businesses']=''
latlng_pd.columns=['Indice','CVE_ENT','CVE_MUN','CVE_AGEB','lat','lon']#,'Number of near Businesses'
del latlng_pd['Indice']
latlng_pd=latlng_pd.iloc[0:10,:]
latlng_pd['CVE_AGEB'] = latlng_pd['CVE_AGEB'].str.zfill(4)
latlng_pd
```

```python
target_search=input("What you want to search: ")
target_type=input("What is your business type: ")
```

```python

# geocoordinates
target_coordinates = ""#19.2762739, -99.1362978"
target_radius = 3000


# set up a parameters dictionary
params = {
    "location": target_coordinates,
    "keyword": target_search,
    "radius": target_radius,
    "type": target_type,
    "key": gkey
}

# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# run a request using our params dictionary
#response = requests.get(base_url, params=params)
```

```python
# convert response to json
#places_data = response.json()
#len(places_data['results'])
# Print the json (pretty printed)
#print(json.dumps(places_data, indent=4, sort_keys=True))
```

```python
for index, row in latlng_pd.iterrows():
    # get lat, lng from df
    lat = row["lat"]
    lng = row["lon"]
    # change location each iteration while leaving original params in place
    params["location"] = f"{lng},{lat}"
    # make request and print url
    response = requests.get(base_url, params=params)
    places_data = response.json()
    latlng_pd.loc[index, "Number of near Business"] = len(places_data["results"])
```
