```python
import pandas as pd
import json
```

```python
state_f = "CDMX&MetroDF.csv"
rezago_f = "RSocial_AGEB 2010_v2.csv"

read_state = pd.read_csv(state_f, encoding='utf-8')
read_state = read_state.loc[read_state['NOM_LOC'] == 'Total AGEB urbana']
read_state
```

```python
read_rezago = pd.read_csv(rezago_f, encoding='utf-8')
read_rezago = read_rezago.rename(columns={'CVE_AGEB' : 'AGEB'})
read_rezago = read_rezago.loc[(read_rezago['Entidad federativa'] == 'Distrito Federal') | (read_rezago['CVE_ENT'] == 15)]
read_rezago
```

```python
merged_df = read_state.merge(read_rezago, on="AGEB")
merged_df
```
