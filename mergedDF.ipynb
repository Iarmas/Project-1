```python
import pandas as pd
```

```python
state_f = "CDMX&MetroDF.csv"
rezago_f = "RSocial_AGEB 2010_v2.csv"

read_state = pd.read_csv(state_f, encoding='utf-8')
read_state = read_state.loc[read_state['NOM_LOC'] == 'Total AGEB urbana']
read_state['AGEB'] = read_state['AGEB'].str.zfill(4)
read_state.head()
```

```python
read_rezago = pd.read_csv(rezago_f, encoding='utf-8')
read_rezago = read_rezago.rename(columns={'CVE_AGEB' : 'AGEB', 'CVE_MUN' : 'MUN', 'CVE_LOC' : 'LOC', 'CVE_ENT' : 'ENTIDAD'})
read_rezago = read_rezago.loc[(read_rezago['Entidad federativa'] == 'Distrito Federal') | (read_rezago['ENTIDAD'] == 15)]
read_rezago['MUN'] = read_rezago['MUN'].astype(int)
read_rezago.head()
```

```python
merged_df = pd.merge(read_state, read_rezago, on=["AGEB", 'ENTIDAD', 'MUN', 'LOC'])
merged_df
```

```python
merged_df.to_csv("GranTablaCDMX&Metro.csv")
```
