```python
import pandas as pd
```

```python
ageb_rez = pd.read_csv("RSocial_AGEB 2010_v2.csv")

cdmx = pd.read_csv("Estados/Total_CDMX.csv")
cdmx2 = pd.read_csv("Estados/Total_CDMX2.csv")
```

```python
ageb_rez.head()
```

```python
ageb_rez_df = ageb_rez.loc[ageb_rez['Entidad federativa'] == 'Distrito Federal'].copy()
ageb_rez_df = ageb_rez_df.rename(columns={'CVE_AGEB' : 'AGEB'})
ageb_rez_df.head()
```

```python
cdmxTot = cdmx.append(cdmx2, ignore_index=True)
cdmxTot.head()
```

```python
cdmxTot_ageb = cdmxTot.loc[cdmxTot['NOM_LOC'] == 'Total AGEB urbana'].copy()
```

```python
merged = pd.merge(cdmxTot_ageb, ageb_rez_df, on='AGEB').set_index('AGEB')
merged.head()
```
