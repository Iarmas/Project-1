```python
%matplotlib notebook

import csv
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from functools import reduce

GT = 'GranTablaCDMX&Metro.csv'
GT_df = pd.read_csv(GT,encoding="utf-8")
GT_df
```

```python
GT_df.reset_index()

GT_df['P_0A2'] = GT_df['P_0A2'].str.replace('*', '0')

edad_0a2_MUN = GT_df.loc[:,['NOM_ENT','NOM_MUN','AGEB','P_0A2','POBTOT']]
edad_0a2_MUN

edad_0a2_MUN['Porcentaje Pob. 0 a 2 años'] = edad_0a2_MUN['P_0A2'].astype(float)/edad_0a2_MUN['POBTOT'].astype(float)*100
edad_0a2_MUN.head()
```
